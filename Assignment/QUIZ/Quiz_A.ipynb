## 1. Classification

  Gunakan dataset **breast-cancer-wisconsin**, dan hitung accuracy, F1, precision, dan recall dari minimal **3 algoritma**!

  Silahkan merujuk ke file **.names** untuk melihat attributenya! Dipersilahkan untuk memproses nama feature menggunakan pandas atau mengubah langsung file .csv nya (ditambahkan header/judul per kolomnya)

  **Jelaskan segala hal yang dilakukan** (preprocessing, hyperparameter tuning, dll) dengan sejelas mungkin!


```
#Import Libray yang digunakan

import os
import numpy as np
import pandas as pd
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
#import dataset yang akan digunakan
dataset = pd.read_csv('/content/drive/My Drive/Datasets/breast-cancer-wisconsin.csv')
```

```
#Melihat bbrapa data awal dan type data tiap kolom nya
# dataset.to_csv("breast-cancer-wisconsin.csv", header=["Clump_thickness", "Uniformity_of_cell_size", "Uniformity_of_cell_shape", "Marginal_adhesion", "Single_epithelial_cell_size", "Bare_nuclei", "Bland_chromatin", "Nomral_nucleoli", "Mitoses", "Class"], index=False)
dataset.head()
dataset.dtypes
```

```
#Library yang digunakan untuk preprocessing dan algoritma
from sklearn.ensemble import  RandomForestClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```
feature = dataset.drop(['Class', 'Bare_nuclei'], axis=1) #Class di drop karna merupakan target yang akan diprediksi, Bare_nuclei di drop karena bertype data string
label = dataset['Class'] #target prediksi
```

```
#proces preprocessing dsni saya menggunakan standardScaler()
scaler = preprocessing.StandardScaler()

feature_scaled = scaler.fit_transform(feature.values) 
```

```
#Random Forest Classifier
#disini saya menggunakan Random Forest Classifier dengan n_estimators 800 dan max_features sqrt
feature_train, feature_test, label_train, label_test = train_test_split(feature_scaled, label, test_size=0.2, random_state=10)
random_forest = RandomForestClassifier(n_estimators=800, bootstrap=True, max_features = "sqrt")
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)

print("prediksi menggunakan training set")
print('Akurasi dari Testing Set: {:.3f}'.format(random_forest.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_train, predictions, average='macro'))
print("precision score")
print(precision_score(label_train, predictions, average='macro'))
print("recall score")
print(recall_score(label_train, predictions, average='macro'))

predictions = random_forest.predict(feature_test)

print("\nprediksi menggunakan testing set")
print('Akurasi dari Testing Set: {:.3f}'.format(random_forest.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_test, predictions, average='macro'))
print("precision score")
print(precision_score(label_test, predictions, average='macro'))
print("recall score")
print(recall_score(label_test, predictions, average='macro'))
```

```
#GradienBoosting
#disini saya menggunakan n_estimators 50 selebihnya default
from sklearn.ensemble import GradientBoostingClassifier
GB_classi = GradientBoostingClassifier(n_estimators=50)
GB_classi.fit(feature_train, label_train)


predictions = GB_classi.predict(feature_train)

print("prediksi menggunakan training set")
print('Akurasi dari Training Set: {:.3f}'.format(GB_classi.score(feature_train, label_train))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_train, predictions, average='macro'))
print("precision score")
print(precision_score(label_train, predictions, average='macro'))
print("recall score")
print(recall_score(label_train, predictions, average='macro'))

predictions = GB_classi.predict(feature_test)

print("\nprediksi menggunakan testing set")
print('Akurasi dari Testing Set: {:.3f}'.format(GB_classi.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_test, predictions, average='macro'))
print("precision score")
print(precision_score(label_test, predictions, average='macro'))
print("recall score")
print(recall_score(label_test, predictions, average='macro'))
```

```
#k-NN
#untuk KNN saya menggunakan nilai k 21
knn = KNeighborsClassifier(21)
knn.fit(feature_train, label_train)

predictions = knn.predict(feature_train)

print("prediksi menggunakan training set")
print('Akurasi dari Testing Set: {:.3f}'.format(knn.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_train, predictions, average='macro'))
print("precision score")
print(precision_score(label_train, predictions, average='macro'))
print("recall score")
print(recall_score(label_train, predictions, average='macro'))

predictions = knn.predict(feature_test)

print("\nprediksi menggunakan testing set")
print('Akurasi dari Testing Set: {:.3f}'.format(knn.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_test, predictions, average='macro'))
print("precision score")
print(precision_score(label_test, predictions, average='macro'))
print("recall score")
print(recall_score(label_test, predictions, average='macro'))
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```
dataset2 = pd.read_csv('/content/drive/My Drive/Datasets/banknote.csv')
```

```
dataset2.head()
```

```
dataset2.dtypes
```

```
feature2 = dataset2.drop(['class'], axis=1) #Class di drop karna merupakan target yang akan diprediksi
label2 = dataset2['class'] #target prediksi
```

```
#disini saya masih menggunakan StandarScaler untuk proces preprocessing nya
scaler = preprocessing.StandardScaler()

feature_scaled = scaler.fit_transform(feature2.values) 
feature_train, feature_test, label_train, label_test = train_test_split(feature_scaled, label2, test_size=0.2, random_state=10)
```

```
#Random Forest Regressor
#disini saya menggunakan parameter yang sama dengan Random Forest Classifier yaitu n_estimators 800 max_features sqrt sisa nya itu default
random_forest = RandomForestRegressor(n_estimators=800, bootstrap=True, max_features = "sqrt",max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_leaf_nodes=None)
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)

print("prediksi menggunakan training set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_train, predictions))
print("MAE")
print(mean_absolute_error(label_train, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_train, predictions)))

predictions = random_forest.predict(feature_test)

print("\nprediksi menggunakan testing set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_test, predictions))
print("MAE")
print(mean_absolute_error(label_test, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_test, predictions)))
```

```
#XGBoost
#pada XGBoost saya menggunakan parameter max_depth 3, learning rate nya 0.02, n_estimators 600, objective reg:squarederrr, dan booster gbtree
xgb_model = xgb.XGBRegressor(max_depth=3, learning_rate=0.02, n_estimators=600, verbosity=1, objective='reg:squarederror', booster='gbtree',)
xgb_model.fit(feature_train, label_train)

predictions = xgb_model.predict(feature_train)

print("prediksi menggunakan training set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_train, predictions))
print("MAE")
print(mean_absolute_error(label_train, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_train, predictions)))

predictions = xgb_model.predict(feature_test)

print("\nprediksi menggunakan testing set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_test, predictions))
print("MAE")
print(mean_absolute_error(label_test, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_test, predictions)))
```

```
#KNN Regressor
#untuk KNN ini saya menggunakan nilai k nya 19 dan leaf_size nya 30 sisa nya itu default
knn = KNeighborsRegressor(n_neighbors=19, leaf_size=30, p=2, n_jobs=None, algorithm='auto')
knn.fit(feature_train, label_train)

predictions = knn.predict(feature_train)

print("prediksi menggunakan training set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_train, predictions))
print("MAE")
print(mean_absolute_error(label_train, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_train, predictions)))

predictions = knn.predict(feature_test)

print("\nprediksi menggunakan testing set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_test, predictions))
print("MAE")
print(mean_absolute_error(label_test, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_test, predictions)))
```
