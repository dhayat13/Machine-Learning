```
dataset = pd.read_csv('/content/drive/My Drive/Datasets/horse-colic.csv')
```

```
dataset.columns = ['surgery','age','hosp_numb','rect_temp','pulse','resp_rate','temp_ext','perph_pulse','mucous','capill','pain','perist','abd_dist','nas_tub','nas_ref','nas_ref_ph','rect_ex','abdoment','pcv','tot_pro','abd_app','abd_tot_pro','outcome','surg_less','type_less_a','type_less_b','type_less_c','cp_data']
```

```
dataset.replace("?", np.nan, inplace=True) #digunakan untuk mengganti data yang nilai nya ? menjadi NaN menggunakan fungsi dari numpy
```

```
dataset.head()
```

```
dataset.isnull().sum().sort_values() #untuk mengatahui apakah ada nilai Null dan di urutkan agar memudahkan
```

```
dataset.drop(['rect_ex', 'nas_tub', 'nas_ref', 'abdoment', 'abd_app', 'abd_tot_pro', 'nas_ref_ph'], axis=1, inplace=True)
```

```
# dataset.head()
dataset.isnull().sum().sort_values()
```

```
dataset.dropna(subset=["outcome","surgery","capill","perist","mucous","pain","abd_dist","temp_ext","perph_pulse"], axis=0, inplace=True)
```

```
avg_rect_temp = dataset['rect_temp'].astype('float').mean(axis=0)
avg_resp_rate = dataset['resp_rate'].astype('float').mean(axis=0)
avg_pcv = dataset['pcv'].astype('float').mean(axis=0)
avg_tot_pro  = dataset['tot_pro'].astype('float').mean(axis=0)
avg_pulse = dataset['pulse'].astype('float').mean(axis=0)
```

```
# avg_rect_temp
# avg_pcv
# avg_pulse
# avg_resp_rate
# avg_tot_pro
dataset['rect_temp'].replace(np.nan, avg_rect_temp, inplace=True)
dataset['resp_rate'].replace(np.nan, avg_resp_rate, inplace=True)
dataset['pcv'].replace(np.nan, avg_pcv, inplace=True)
dataset['tot_pro'].replace(np.nan, avg_tot_pro, inplace=True)
dataset['pulse'].replace(np.nan, avg_pulse, inplace=True)
```

```
dataset.isnull().sum().sort_values()
```

```
dataset.dtypes
```

```
dataset['surgery'].value_counts()
```

```
surgery_categorized = pd.get_dummies(dataset['surgery'])
surgery_categorized
```

```
dataset = pd.concat([dataset, surgery_categorized], axis=1)

dataset.head()
```

```
dataset['temp_ext'].value_counts()
```

```
temp_ext_categorized = pd.get_dummies(dataset['temp_ext'])
dataset = pd.concat([dataset, temp_ext_categorized], axis=1)

dataset.head()
```

```
dataset['perph_pulse'].value_counts()
```

```
perph_pulse_categorized = pd.get_dummies(dataset['perph_pulse'])
dataset = pd.concat([dataset, perph_pulse_categorized], axis=1)

dataset.head()
```

```
dataset['mucous'].value_counts()
```

```
mucous_categorized = pd.get_dummies(dataset['mucous'])
dataset = pd.concat([dataset, mucous_categorized], axis=1)

dataset.head()
```

```
dataset['capill'].value_counts()
```

```
capill_categorized = pd.get_dummies(dataset['capill'])
dataset = pd.concat([dataset, capill_categorized], axis=1)
```

```
dataset['pain'].value_counts()
```

```
pain_categorized = pd.get_dummies(dataset['pain'])
dataset = pd.concat([dataset, pain_categorized], axis=1)
```

```
dataset['perist'].value_counts()
```

```
perist_categorized = pd.get_dummies(dataset['perist'])
dataset = pd.concat([dataset, perist_categorized], axis=1)
```

```
dataset['abd_dist'].value_counts()
```

```
abd_dist_categorized = pd.get_dummies(dataset['abd_dist'])
dataset = pd.concat([dataset, abd_dist_categorized], axis=1)
```

```
dataset.drop(['surgery', 'temp_ext', 'perph_pulse', 'mucous', 'abd_dist', 'perist', 'pain','capill'], axis=1, inplace=True)
```

```
dataset.dtypes
```

```
dataset['rect_temp'] = pd.to_numeric(dataset['rect_temp'], errors='coerce')
dataset['pulse'] = pd.to_numeric(dataset['pulse'], errors='coerce')
dataset['resp_rate'] = pd.to_numeric(dataset['resp_rate'], errors='coerce')
dataset['pcv'] = pd.to_numeric(dataset['pcv'], errors='coerce')
dataset['tot_pro'] = pd.to_numeric(dataset['tot_pro'], errors='coerce')
dataset['outcome'] = pd.to_numeric(dataset['outcome'], errors='coerce')
```

```
dataset.dtypes
```

```
#Library yang digunakan untuk preprocessing dan algoritma
from sklearn.ensemble import  RandomForestClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```
feature = dataset.drop(['outcome'], axis=1) #Outcome di drop karna merupakan target yang akan diprediksi
label = dataset['outcome'] #target prediksi
```

```
#proces preprocessing dsni saya menggunakan standardScaler()
scaler = preprocessing.StandardScaler()

feature_scaled = scaler.fit_transform(feature.values) 
feature_train, feature_test, label_train, label_test = train_test_split(feature_scaled, label, test_size=0.2, random_state=10)
```

```
#Random Forest Classifier
#disini saya menggunakan Random Forest Classifier dengan n_estimators 500 dan max_features sqrt
random_forest = RandomForestClassifier(n_estimators=500, bootstrap=True, max_features = "sqrt")
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)

print("prediksi menggunakan training set")
print('Akurasi dari Testing Set: {:.3f}'.format(random_forest.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_train, predictions, average='macro'))
print("precision score")
print(precision_score(label_train, predictions, average='macro'))
print("recall score")
print(recall_score(label_train, predictions, average='macro'))

predictions = random_forest.predict(feature_test)

print("\nprediksi menggunakan testing set")
print('Akurasi dari Testing Set: {:.3f}'.format(random_forest.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_test, predictions, average='macro'))
print("precision score")
print(precision_score(label_test, predictions, average='macro'))
print("recall score")
print(recall_score(label_test, predictions, average='macro'))
```

```
#GradienBoosting
#disini saya menggunakan n_estimators 50 selebihnya default
from sklearn.ensemble import GradientBoostingClassifier
GB_classi = GradientBoostingClassifier(n_estimators=50)
GB_classi.fit(feature_train, label_train)


predictions = GB_classi.predict(feature_train)

print("prediksi menggunakan training set")
print('Akurasi dari Training Set: {:.3f}'.format(GB_classi.score(feature_train, label_train))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_train, predictions, average='macro'))
print("precision score")
print(precision_score(label_train, predictions, average='macro'))
print("recall score")
print(recall_score(label_train, predictions, average='macro'))

predictions = GB_classi.predict(feature_test)

print("\nprediksi menggunakan testing set")
print('Akurasi dari Testing Set: {:.3f}'.format(GB_classi.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_test, predictions, average='macro'))
print("precision score")
print(precision_score(label_test, predictions, average='macro'))
print("recall score")
print(recall_score(label_test, predictions, average='macro'))
```

```
#k-NN
#untuk KNN saya menggunakan nilai k 15
knn = KNeighborsClassifier(15)
knn.fit(feature_train, label_train)

predictions = knn.predict(feature_train)

print("prediksi menggunakan training set")
print('Akurasi dari Testing Set: {:.3f}'.format(knn.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_train, predictions, average='macro'))
print("precision score")
print(precision_score(label_train, predictions, average='macro'))
print("recall score")
print(recall_score(label_train, predictions, average='macro'))

predictions = knn.predict(feature_test)

print("\nprediksi menggunakan testing set")
print('Akurasi dari Testing Set: {:.3f}'.format(knn.score(feature_test, label_test))) # .score() -> Accuracy
print("f1 score")
print(f1_score(label_test, predictions, average='macro'))
print("precision score")
print(precision_score(label_test, predictions, average='macro'))
print("recall score")
print(recall_score(label_test, predictions, average='macro'))
```

```
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```
dataset2 = pd.read_csv('/content/drive/My Drive/Datasets/abalone1.csv')
```

```
# dataset.columns = ['Sex','Length','Diameter','Height','Whole_weight','Shucked_weight','Viscera_weight','Shell_weight','Rings']
dataset2.head()
```

```
dataset2.dtypes
```

```
cleanup = {
    'Sex':{
        'M': 1,
        'F' : 2,
        'I' : 3
    },
}
dataset2.replace(cleanup, inplace=True)
```

```
dataset2.dtypes
```

```
feature2 = dataset2.drop(['Sex'], axis=1) #Sex di drop karna merupakan target yang akan diprediksi
label2 = dataset2['Sex'] #target prediksi
```

```
#disini saya masih menggunakan StandarScaler untuk proces preprocessing nya
scaler = preprocessing.StandardScaler()

feature_scaled = scaler.fit_transform(feature2.values) 
feature_train, feature_test, label_train, label_test = train_test_split(feature_scaled, label2, test_size=0.2, random_state=10)
```

```
#Random Forest Regressor
#disini saya menggunakan parameter yang sama dengan Random Forest Classifier yaitu n_estimators 800 max_features sqrt sisa nya itu default
random_forest = RandomForestRegressor(n_estimators=400, bootstrap=True, max_features = "sqrt",max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_leaf_nodes=None)
random_forest.fit(feature_train, label_train)

predictions = random_forest.predict(feature_train)

print("prediksi menggunakan training set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_train, predictions))
print("MAE")
print(mean_absolute_error(label_train, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_train, predictions)))

predictions = random_forest.predict(feature_test)

print("\nprediksi menggunakan testing set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_test, predictions))
print("MAE")
print(mean_absolute_error(label_test, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_test, predictions)))
```

```
#XGBoost
#pada XGBoost saya menggunakan parameter max_depth 3, learning rate nya 0.02, n_estimators 600, objective reg:squarederrr, dan booster gbtree
xgb_model = xgb.XGBRegressor(max_depth=3, learning_rate=0.02, n_estimators=600, verbosity=1, objective='reg:squarederror', booster='gbtree',)
xgb_model.fit(feature_train, label_train)

predictions = xgb_model.predict(feature_train)

print("prediksi menggunakan training set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_train, predictions))
print("MAE")
print(mean_absolute_error(label_train, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_train, predictions)))

predictions = xgb_model.predict(feature_test)

print("\nprediksi menggunakan testing set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_test, predictions))
print("MAE")
print(mean_absolute_error(label_test, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_test, predictions)))
```

```
#KNN Regressor
#untuk KNN ini saya menggunakan nilai k nya 19 dan leaf_size nya 30 sisa nya itu default
knn = KNeighborsRegressor(n_neighbors=19, leaf_size=30, p=2, n_jobs=None, algorithm='auto')
knn.fit(feature_train, label_train)

predictions = knn.predict(feature_train)

print("prediksi menggunakan training set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_train, predictions))
print("MAE")
print(mean_absolute_error(label_train, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_train, predictions)))

predictions = knn.predict(feature_test)

print("\nprediksi menggunakan testing set")
# print('Akurasi dari Testing Set: {:.3f}'.format(lin_reg.score(feature_test, label_test))) # .score() -> Accuracy
print("r2 score")
print(r2_score(label_test, predictions))
print("MAE")
print(mean_absolute_error(label_test, predictions))
print("RMSE")
print(np.sqrt(mean_squared_error(label_test, predictions)))
```
