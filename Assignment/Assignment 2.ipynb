1. Buatlah suatu list kosong bernama `numbers`. Lalu, buatlah loop yang secara otomatis menambahkan nomor 1-10 ke list tersebut. 
  * Hint: Gunakan function `range()` untuk melakukan loop selama `n` kali

2. Tambahkan setiap nilai numbers sebanyak 2.

3. Cek untuk setiap angka ganjil, apabila ganjil tambahkan angka 1.

4. Hapus semua angka yang lebih kecil dari 5 di list.
  * Hint: Gunakan function `remove()`

5. Print semua nilai isi di list dengan format:
  * Angka ke-(index) bernilai (angka)
  * Misal: Angka ke-0 bernilai 10

6. Print semua angka yang habis dibagi 2, dengan format:
  * Angka ke-(index) bernilai (angka)
  * Misal: Angka ke-0 bernilai 10

```
#Soal 1
print("soal 1")
print("Buat suatu List Kosong, Lalu Buat Loop untuk menambahkan isi List secara otomatis")

print("\n*List Kosong")
numbers = []
print(numbers)
for i in range (1, 11):
  numbers.append(i)
print(numbers)
print()

#Soal 2
print("soal 2")
print("Tambahkan Setiap Nilai Numbers Sebanyak 2")
new_numbers = []
for i in numbers:
  i += 2
  new_numbers.append(i)
print(new_numbers)
print()

#soal 3
print("soal 3")
print("Cek Setiap Angka Ganjil, Jika Ganjil di Tambahkan angka 1")
odd_numbers = []
for i in new_numbers:
  if i % 2 != 0:
    odd_numbers.append(i+1)
  else:
    odd_numbers.append(i)
print(odd_numbers)
print()

#soal 4
print("soal 4")
print("Hapus Semua angka yang < 5 di List")
x=0
numbers.sort()
for i in numbers:
  if i < 5:
    x+=1
numbers=numbers[x:]
print(numbers)
print()

#soal 5
print("soal 5")
print("Tampilkan Semua Nilai di List :")
for index, value in enumerate(numbers):
  print("Angka Ke-"+str(index), "Bernilai :"+str(value))

print()

#soal 6
print("soal 6")
print("Semua Angka yang habis dibagi 2 :")
for index, value in enumerate(numbers):  
  if value % 2 == 0:
    print("Angka Ke-"+str(index), "Bernilai :"+str(value))

print()
```
