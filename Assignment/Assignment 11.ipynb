**2. ALGORITMA FILTERRING**

```
#Average Filterring
kernel = [3, 5, 7, 9]
for i in kernel:
  for j in data:
    print(i ,' x ', i)
    img_avg = cv.blur(j, (i, i)) 
    cv2_imshow(img_avg)
    print()
```

```
#Gaussian Filter
for i in kernel:
  for j in data:
    print(i ,' x ', i)
    img_gaus = cv.GaussianBlur(j, (i, i), 0) 
    cv2_imshow(img_gaus)
    print()
```

```
#Median Filter
for i in kernel:
  for j in data:
    print(i ,' x ', i)
    img_median = cv.medianBlur(j, i) 
    cv2_imshow(img_median)
    print()
```

**Kesimpulan**

Untuk gambar kedua (image2 / phone noise 1) gambar terbaik yaitu hasil dari median filter kernel 3x3 disana dapat terlihat noise pada gambar tersebut sudah tidak terlalu kelihatan, namun untuk kernel 5x5 nya gambar nya terlalu blur.


---

Unutk gambar ketiga (image3 / phone noise 2) gambar terbaik yaitu hasil dari median filter kernel 5x5 karena dapat dilihat noise nya sudah sangat jauh berkurang.

**3. THRESHOLDING**

```
phone_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
phone_noise_gray = cv.cvtColor(image2, cv.COLOR_BGR2GRAY)
phone_noise2_gray = cv.cvtColor(image3, cv.COLOR_BGR2GRAY)
phone_salt_gray = cv.cvtColor(image4, cv.COLOR_BGR2GRAY)
phone_salt2_gray = cv.cvtColor(image5, cv.COLOR_BGR2GRAY)

cv2_imshow(phone_gray)
cv2_imshow(phone_noise_gray)
cv2_imshow(phone_noise2_gray)
cv2_imshow(phone_salt_gray)
cv2_imshow(phone_salt2_gray)
```

```
#GLOBAL TRHESHOLDING

img_list = [phone_gray, phone_noise_gray, phone_noise2_gray, phone_salt_gray, phone_salt2_gray]
img_id = ['phones', 'phones-noise-1', 'phone-noise-2', 'phones-salt-pepper-1', 'phones-salt-pepper-2']

# def globalthresh(image_name, image_id):
#   threshold_value, threshold_result = cv.threshold(image_name, 127, 255, cv.THRESH_BINARY) # Binary thresholding
#   print('Image Name : ', image_id)
#   print('Global Thresholding Value : ', threshold_value)
#   cv2_imshow(threshold_result)
#   print()

# for i in range(len(img_list)):
#   globalthresh(img_list[i], img_id[i])

for i in range(len(img_list)):
  thresholding_value, thresholding_result = cv.threshold(img_list[i], 127, 255, cv.THRESH_BINARY)
  print('Image Name : ', img_id[i])
  print('Global Thresholding Value : ', thresholding_value)
  cv2_imshow(thresholding_result)
  print()
```

```
#OTSU'S THRESHOLDING
# def otsuthresh(image_name, image_id):
#   thresholding_value, thresholding_result = cv.threshold(image_name, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
#   print('Image Name : ', image_id)
#   print('Otsu Thresholding Value : ', thresholding_value)
#   cv2_imshow(thresholding_result)
#   print()

# for i in range(len(img_list)):
#   otsuthresh(img_list[i], img_id[i])

for i in range(len(img_list)):
  thresholding_value, thresholding_result = cv.threshold(img_list[i], 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
  print('Image Name : ', img_id[i])
  print('Otsu Thresholding Value : ', thresholding_value)
  cv2_imshow(thresholding_result)
  print()
```

```
#ADAPTIVE THRESHOLDING WITH MEAN

# def adaptivethresh(image_name, image_id):
#   threshold_result = cv.adaptiveThreshold(image_name, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
#   print('Image Name : ', image_id)
#   cv2_imshow(threshold_result)
#   print()

# for i in range(len(img_list)):
#   adaptivethresh(img_list[i], img_id[i])

for i in range(len(img_list)):
  thresholding_result = cv.adaptiveThreshold(img_list[i], 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
  print('Image Name : ', img_id[i])
  cv2_imshow(thresholding_result)
  print()
```

```
#ADAPTIVE THRESHOLDING WITH GAUSSIAN

# def adaptivethresh(image_name, image_id):
#   threshold_result = cv.adaptiveThreshold(image_name, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
#   print('Image Name : ', image_id)
#   cv2_imshow(threshold_result)
#   print()

# for i in range(len(img_list)):
#   adaptivethresh(img_list[i], img_id[i])

for i in range(len(img_list)):
  thresholding_result = cv.adaptiveThreshold(img_list[i], 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
  print('Image Name : ', img_id[i])
  cv2_imshow(thresholding_result)
  print()
```

**KESIMPULAN**

---

Untuk Global Thresholding hasil nya kurang baik, karena masih terbayang bayang hitam yang disebabkan metode global thresholding ini langsung memotong di nilai thresholding 127.

---

Untuk Otsu's Thresholding hasil nya agak lebih baik dari global thresholding, karena nilai thresholding nya otomatis dari algoritma nya.

---

Untuk Adaptive Thresholding menggunakan Gaussian hasil nya lebih baik karena gambar dari handphones nya lebih jelas, namun dsni masih dapat dilihat juga background nya.
